```python
# Lib untuk manipulasi data
import numpy as np
import pandas as pd

# lib untuk praproses dataset
from sklearn.preprocessing import MinMaxScaler
```

```python
# load dataset
dataset = pd.read_csv("dataset/ekstrasi-fitur-kopra-nonTelungkup_v2.csv")
```

```python
# show metadata
dataset.info()
```

```python
dataset.describe()
```

```python
# normalize features
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(dataset)
```

```python
df = pd.DataFrame(
  np.round(scaled,4),
  columns=[
    # color
    "Mean_R", "Mean_G", "Mean_B", "Mean_H", "Mean_V", "Mean_S", "Mean_Gray", "Standar_Deviasi", 
    # shape
    "Luas", "Perimeter",
    # texture
    "Contrast", "Dissimilarity", "Homogeneity", "Energy", "Correlation", "Jenis_Kopra"
  ]
)
```

```python
df.to_csv("dataset/ekstrasi-fitur-kopra-nonTelungkup_norm.csv", index=False)
```
